```python
import pandas as pd
import matplotlib
from matplotlib import pyplot as plt
from pylab import rcParams
from datetime import date
```

```python
real = pd.read_csv("real_price.csv")
est_df = pd.read_csv("xgb_price.csv")
moving_avg = pd.read_csv("moving_avg_price.csv")
```

```python
# Plot adjusted close over time, only for test set
rcParams['figure.figsize'] = 10, 8 # width 10, height 8
matplotlib.rcParams.update({'font.size': 14})

ax = real.plot(x='Date', y='adj_close', style='gx-', grid=True)
ax = moving_avg.plot(x='date', y='est_N2', style='bx-', grid=True, ax=ax)
ax = est_df.plot(x='Date', y='est', style='rx-', grid=True, ax=ax)

ax.legend(['real price', 'moving average', 'xgboost'], loc='upper left')
ax.set_xlabel("Date")
ax.set_ylabel("$GOOGL stock price")
# ax.set_xlim([date(2019, 11, 1), date(2020, 3, 1)])
ax.set_ylim([1280, 1500])
```

```python
ax.figure.savefig("test.png")
```
